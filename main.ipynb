# Read data

```python
import csv
import os

# open and read .csv
f = open(f"{os.getcwd()}\\data\\106_MLII.csv", "r")
data = csv.reader(f)

next(data) # skip header

# Convert type csv_reader to list
# csv_reader --> list
y: list = []
for row in data:
    y.append(int(row[0]))
# close .csv
f.close()

print(y[:10])
```

# Show Graph

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array(y)
plt.figure(figsize=(100, 20))
plt.plot(y[:1000])
plt.show()
```

# Find Peak

```python
from scipy.signal import find_peaks

# y over 1250
peaks, _ = find_peaks(y, height=(1200, 1600), distance=200)
plt.figure(figsize=(100, 20), frameon=False)
plt.plot(y[:2000])
plt.plot(peaks[:2], y[:2000][peaks[:2]], "x")
# plt.savefig(f"{os.getcwd()}\\train_data\\1")
print(len(peaks))
plt.show()
```

# Split Figure

```python
print(len(peaks))
for pulse_N in range(1, len(peaks)-1):
    fig = plt.figure(frameon=False)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.axis("off")
    ax.plot(y[peaks[pulse_N]: peaks[pulse_N+1]])
    # plt.figure(figsize=(2, 2), frameon=False)
    # plt.plot(y[peaks[pulse_N]: peaks[pulse_N+1]])
    # plt.savefig(f"{os.getcwd()}\\train_data\\{pulse_N}")
    plt.close()
```
